```python
# Libraries Importing
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts teh configuration of the plots we will create
```

```python
import numpy as np
```

```python
# Read in the Data
df = pd.read_csv('/Users/emekaudemezue/Downloads/Data Analytics/Data Analyst Portfolio/Video 4/movies.csv')
```

```python
df.head()
```

```python
# Searching for missing data
for col in df.columns:
    pct_miss = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_miss))
```

```python
# Replacing missing/null values
budget_mean = df['budget'].mean()
df['budget'].replace(np.nan,budget_mean, inplace=True)
df.head()

gross_mean = df['gross'].mean()
df['gross'].replace(np.nan,gross_mean, inplace=True)
df.head()
```

```python
# Column Datatypes
df.dtypes
```

```python
# Changing 'Budget' column datatype to 'int64'
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create Correct Year Column
df['yearcorrect'] = df['released'].astype(str).str[9:13]
df
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# Dropping any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df
```

```python
# Budget High Correlation
# Company Name High Correlation
# Build scatter plot to correlate budget against gross
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs. Gross Revenue')
plt.xlabel("Gross Revenue")
plt.ylabel("Movie Budget")
plt.show
```

```python
# Regression Plot Creation using Seaborn
sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':"black"}, line_kws={"color":"red"})
```

```python
# Correlation calculation (Methods: 'Pearson','Kendall','Spearman')
df.corr(method='pearson')
```

```python
# High Correlation between Budget and Gross
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features ')
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features ')
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr()
correlation_pairs = correlation_matrix.unstack()
correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
