```python
y = athletes.NOC.value_counts().values[0:31]
x = athletes.NOC.value_counts().index[0:31]
plt.figure(figsize=(30,7))
plt.bar(x,y)
plt.xlabel('Country')
plt.ylabel('Number of Athletes')
plt.xticks(rotation='vertical',size=10)
plt.yticks(size=15)
plt.show()
```

```python
y = athletes.Discipline.value_counts().values[0:31]
x = athletes.Discipline.value_counts().index[0:31]
plt.figure(figsize=(20,5))
plt.bar(x,y)
for index, value in enumerate(y):
    plt.text(index, value, str(value),color='blue',size=10,fontweight='bold')
plt.xlabel('Country')
plt.ylabel('Disciplines')
plt.xticks(rotation='vertical',size=10)
plt.yticks(size=15)
plt.show()
```

```python
# Athletics has the highest number of participants with a number of 2068 followed with Swimmming.
```

```python
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in athletes.Name)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=300, background_color="white").generate(text)
plt.figure(figsize=(30,6))
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
